```python
#operacion con listas
#ian dominguez

frutas = ['manzana', 'pera','uva','mango']

print(frutas[0])
print(frutas[-1])

frutas[1] = 'naranja'
frutas.append('kiwi')
frutas.insert(1, 'fresa') # Fixed: added comma
frutas.remove('uva')    # Fixed: corrected 'fruas'
frutas.pop(0)           # Fixed: corrected 'frutas,pop(0)'
```

```python
#lista con numeros
#ian dominguez

numeros = [45, 12, 17, 78, 3, 56, 89, 23, 67, 1, 34]
print("original: ", numeros)

# ordenar de menor a mayor
numeros.sort()
print ("ordena:", numeros)

# ordenar de mayor a menor
numeros.reverse()
print("invertida:", numeros)
```

```python
#promedio
#ian dominguez

numeros = [45, 12, 17, 78, 3, 56, 89, 23, 67, 1, 34]

total = sum(numeros)
promedio = sum(numeros) / len(numeros)
print(f"suma: {total}")
print(f"promedio: {promedio:.2f}")
```
